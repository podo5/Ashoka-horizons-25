```python

starbucks.columns = starbucks.columns.str.strip()

plt.figure(figsize=(10, 6))
sns.scatterplot(x=starbucks['Calories'], y=starbucks['Sugars (g)'])
plt.title('Relationship between Calories and Sugars in Starbucks')
plt.xlabel('Calories')
plt.ylabel('Sugars (g)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=PIMA, x='Outcome', y='Glucose', palette='Set3')
plt.title('Distribution of Glucose Levels by Diabetes Outcome')
plt.xlabel('Diabetes Outcome (0 = No, 1 = Yes)')
plt.ylabel('Glucose Level')
plt.show()
```

```python
yearly_counts = netflix['release_year'].value_counts().sort_index()
yearly_counts.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o', color='crimson')
plt.title('Number of Netflix Titles Released Each Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
```

```python
type_year_counts = netflix.groupby(['release_year', 'type']).size().unstack()
type_year_counts.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(type_year_counts.index, type_year_counts['Movie'], label='Movies', marker='o', color='blue')
plt.plot(type_year_counts.index, type_year_counts['TV Show'], label='TV Shows', marker='s', color='green')

plt.title('Number of Netflix Titles Released Each Year by Type')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=PIMA, x='Age', y='BMI', color='purple')
plt.title('Scatter Plot of BMI vs Age')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=PIMA, x='Age', y='BMI', hue='Outcome', palette='Set1')
plt.title('BMI vs Age Colored by Diabetes Outcome')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.legend(title='Diabetes Outcome')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.scatterplot(
    data=PIMA,
    x='Age',
    y='BMI',
    hue='Outcome',
    size='Pregnancies',
    palette='Set2',
    alpha=0.7,
    sizes=(20, 200)  # minimum and maximum point sizes
)

plt.title('BMI vs Age Colored by Diabetes Outcome and Sized by Pregnancies')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.legend(title='Outcome & Pregnancies', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
features = ['Glucose', 'BMI', 'Age', 'Pregnancies', 'BloodPressure']
corr_matrix = PIMA[features].corr()
print(corr_matrix)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Selected PIMA Features')
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
axs[0, 0].hist(PIMA['BloodPressure'].dropna(), bins=20, color='orchid', edgecolor='black')
axs[0, 0].set_title('Blood Pressure Distribution')
axs[0, 0].set_xlabel('Blood Pressure')
axs[0, 0].set_ylabel('Frequency')


axs[0, 1].scatter(starbucks['Calories'], starbucks['Protein (g)'], color='skyblue', edgecolors='black')
axs[0, 1].set_title('Calories vs Protein (g)')
axs[0, 1].set_xlabel('Calories')
axs[0, 1].set_ylabel('Protein (g)')


sns.countplot(data=starbucks, y='Beverage_category', ax=axs[1, 0], palette='Set2')
axs[1, 0].set_title('Count of Beverage Categories')
axs[1, 0].set_xlabel('Count')
axs[1, 0].set_ylabel('Beverage Category')


sns.boxplot(data=PIMA, x='Outcome', y='Age', palette='coolwarm', ax=axs[1, 1])
axs[1, 1].set_title('Age Distribution by Diabetes Outcome')
axs[1, 1].set_xlabel('Outcome (0 = No, 1 = Yes)')
axs[1, 1].set_ylabel('Age')


fig.suptitle('Multiview Analysis: Starbucks and PIMA Datasets', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96]) 
plt.show()
```
