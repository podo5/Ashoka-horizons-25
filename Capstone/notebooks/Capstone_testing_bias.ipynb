```python
import math
import random

bias = 0.6  #

A = 1000
B = 1000
i = 10
y_preds = [1000]
x_preds = [1000]
tails = 0
heads = 0

while i > 0:
    i -= 1
    if A <= 0 or B <= 0:
        break
    betAmount = min(A, B) * 0.2
    # Simulating a biased coin flip
    coin_flip = 0 if random.random() < bias else 1
    if coin_flip == 1:  # 0 for heads, 1 for tails
        tails += 1
        B += betAmount
        A -= betAmount
    else:
        heads += 1
        A += betAmount
        B -= betAmount
    y_preds.append(B)
    x_preds.append(A)

print("A: ", x_preds, "B: ", y_preds)
print("Heads: ", heads, "Tails: ", tails)
```

```python
import math
import random
import matplotlib.pyplot as plt

bias = 0.508  # Bias towards heads (Player A advantage)
num_outer = [10, 50, 100, 150, 200]
num_inner = 1000  # More simulations per set

for outer in num_outer:
    A = 1000
    B = 1000
    x_preds = [A]
    y_preds = [B]
    coin_flips = 0
    heads = 0
    tails = 0

    for _ in range(num_inner):
        if A <= 0 or B <= 0:
            break
        coin_flips += 1
        betAmount = min(A, B) * 0.2

        # Biased coin towards heads (A wins more)
        coin_flip = 0 if random.random() < bias else 1
        if coin_flip == 1:  # Tails: B wins
            tails += 1
            B += betAmount
            A -= betAmount
        else:  # Heads: A wins
            heads += 1
            A += betAmount
            B -= betAmount

        x_preds.append(A)
        y_preds.append(B)

    Head_prob = heads / coin_flips if coin_flips > 0 else 0
    print(f"Outer Loop {outer} | Head Probability: {Head_prob:.3f} | Final A: {A:.2f}, B: {B:.2f}")

    # Plotting
    plt.plot(x_preds, label=f'A - {outer} rounds')
    plt.plot(y_preds, label=f'B - {outer} rounds')

plt.title('Balance Over Time (Biased Coin Flip)')
plt.xlabel('Rounds')
plt.ylabel('Money')
plt.legend()
plt.show()
```
